```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/sample_data/homeprices.csv')
```

```python
df
```

```python
sns.lmplot(x='area',y='price',data=df)
```

```python
X = df[['area']].values
y = df['price'].values
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
model.predict([[2700]])
```

```python
model.coef_
```

```python
model.intercept_
```
